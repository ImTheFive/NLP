```python
# 导入相关库
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# 下载数据 类别从 0-9
digits = load_digits(n_class=10)
X = digits.data
y = digits.target
# 可视化数据
fig = plt.gcf() #获取当前图表
fig.set_size_inches(10,12) #1英寸等于 2.54cm

for i in range(0,25):
    ax = plt.subplot(5,5,i+1)
    ax.imshow(digits.images[i],cmap="binary")
    ax.set_title("label="+str(digits.target[i]))
    ax.set_xticks([]) #不显示x轴
    ax.set_yticks([]) #不显示y轴
```

```python
# 随机划分训练集和测试集 测试机比例为0.3
X_train, X_test, y_train, y_test = train_test_split(
                                   X, y, test_size=0.3, random_state=22)
# 使用牛顿迭代法求得系数
clf = LogisticRegression(solver="newton-cg").fit(X_train, y_train)

# 获取得分
clf.score(X_test, y_test)
```

```python
import numpy as np
y_predict = np.array(clf.predict(X_test))
y_test = np.array(y_test)
index = np.arange(0,540)

# 获取预测错误的数据
arr = index[y_predict != y_test]
```

```python
# 显示预测错误的图片

fig = plt.gcf() #获取当前图表
fig.set_size_inches(10,12) #1英寸等于 2.54cm
for i in range(0,arr.shape[0]):
    ax = plt.subplot(5,5,i+1)
    ax.imshow(digits.images[arr[i]],cmap="binary")
    str_title = "label="+str(digits.target[arr[i]])+";predict="+str(y_predict[arr[i]])
    ax.set_title(str_title)
    ax.set_xticks([]) #不显示x轴
    ax.set_yticks([]) #不显示y轴
```
